## Data Cleaning

```python
df.rename(columns={
    'Date (dd/mm/yyyy)':'Date',
    'Transaction (thousands)':'Transaction',
    'Balance (thousands)':'Balance',
    'Bank Received ':'Bank_Received',
    'Customer ID':'Customer_ID',
    'Age Groups':'Age_Groups',
    'City Size (km square)':'City_Size'
    
},inplace=True)
```

```python
df
```

```python
df
```

```python
df['Date']=pd.to_datetime(df['Date'],dayfirst=True,errors='coerce')
```

```python
df['Date'].dtype
```

```python
df.dtypes
```

```python
# Extract required feature like day,date,week,month and year from date
df['Year']=df['Date'].dt.year
df['Month']=df['Date'].dt.month
df['Day']=df['Date'].dt.day
df['Weekday']=df['Date'].dt.day_name()
df['Week']=df['Date'].dt.isocalendar().week

```

```python
df
```

```python
df['Time']=pd.to_datetime(df['Time'],format='%H:%M:%S')
```

```python
df.isna().sum()
```

## EDA

```python
categorical=[
    'Sent From Bank',
    'City',
    'Gender',
    'Age_Groups',
    'Weekday',
    'Age',
    'Bank_Received'
]
```

```python
continuous=[
    'Transaction',
    'Balance',
    'Age'
]
```

```python
time_series=[
    'Year',       
    'Month',             
    'Day',                          
    'Week',
    'Date',    
    'Time'
]
```

```python
df[continuous].describe()
```

```python
for col in continuous:
    sns.histplot(df[col],kde=True,color='teal')
    plt.title(f'{col} Distribution')
    plt.show()
```

```python
for col in categorical:
    count=df[col].value_counts()
    print(f'\n{col}:\n',count)

    sns.countplot(data=df,x=col,palette='viridis')
    plt.title(f'{col} Distribution')
    plt.xticks(rotation=45)
    plt.show()
```

```python
print(df.groupby('Age_Groups')['Transaction'].mean())
print(df.groupby('Gender')['Transaction'].mean())
print(df.groupby('City')['Transaction'].mean())

sns.boxplot(x='Age_Groups',y='Transaction',data=df)
```

```python
sns.boxplot(x='City',y='Transaction',data=df)
```

```python
sns.boxplot(x='Gender',y='Transaction',data=df)
```

```python
df.groupby('Time')['Transaction'].mean().plot(kind='bar',color='coral')
plt.ylabel('Average Transaction Value')
plt.show()
```

```python
sns.heatmap(df[continuous].corr(),annot=True,cmap='coolwarm')
plt.title('Correlation Matrix')
```

```python
daily=df.groupby('Date')['Transaction'].sum()
daily.plot(figsize=(10,4),title='Daily total Transactions')

monthly=df.groupby(['Ye])['Transaction'].sum()
monthly.plot(figsize=(10,4),title='Daily total Transactions')
```

```python
sns.scatterplot(x='Age',y='Balance',hue='Gender',data=df)
plt.title('Balance vs Age by Gender')
```
